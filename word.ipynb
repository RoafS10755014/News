```python
# 本程式碼參考來自： https://clay-atlas.com/blog/2020/01/17/python-chinese-tutorial-gensim-word2vec/
from gensim.corpora import WikiCorpus
import jieba
wiki_corpus = WikiCorpus('/content/drive/MyDrive/zhwiki-20210301-pages-articles-multistream.xml.bz2', dictionary={})
text_num = 0
with open('wiki_text.txt', 'w', encoding='utf-8') as f:
  for text in wiki_corpus.get_texts():
    f.write(' '.join(text)+'\n')
    text_num += 1
    if text_num % 10000 == 0:
      print('{} articles processed.'.format(text_num))

print('{} articles processed.'.format(text_num))
```

用結巴進行斷詞

```python
# coding: utf-8
import jieba
#from opencc import OpenCC

# Initial
#cc = OpenCC('s2t')


# Tokenize
with open('wiki_text_seg.txt', 'w', encoding='utf-8') as new_f:
    with open('/content/drive/MyDrive/wiki_text.txt', 'r', encoding='utf-8') as f:
        for times, data in enumerate(f, 1):
            print('data num:', times)
            #data = cc.convert(data)
            data = jieba.cut(data)
            data = [word for word in data if word != ' ']
            data = ' '.join(data)

            new_f.write(data)
```

訓練模型

```python
from gensim.models import word2vec


# Settings
seed = 666
sg = 0
window_size = 10
vector_size = 100
min_count = 1
workers = 8
epochs = 5
batch_words = 10000

train_data = word2vec.LineSentence('/content/drive/MyDrive/wiki_text_seg.txt')
model = word2vec.Word2Vec(
    train_data,
    min_count=min_count,
    size=vector_size,
    workers=workers,
    #epochs=epochs,
    window=window_size,
    sg=sg,
    seed=seed,
    batch_words=batch_words)

model.save('word2vec.model')
```

開始測試模型訓練結果

```python
from gensim.models import word2vec

model = word2vec.Word2Vec.load('/content/drive/MyDrive/word2vec.model')
print(model.wv['新聞'].shape)

for item in model.wv.most_similar('新聞'):
    print(item)
```
